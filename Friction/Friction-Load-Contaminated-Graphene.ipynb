```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
LoadFresh = pd.read_table('load_fresh.txt')
UnloadFresh = pd.read_table('unload_fresh.txt')
LoadCont = pd.read_table('load_contaminated.txt')
UnloadCont = pd.read_table('unload_contaminated.txt')
```

```python
UnloadCont
```

```python
a_font = {'fontname' : 'Arial'}
```

```python
fig1, ax = plt.subplots()
ax.errorbar(LoadFresh.load, LoadFresh.friction, yerr=LoadFresh.err, fmt='ko', capsize=3, label='Grafeno limpo (carga)')
ax.errorbar(UnloadFresh.load, UnloadFresh.friction, yerr=UnloadFresh.err, fmt='ko', capsize=3, fillstyle='none', label='Grafeno limpo (descarga)')
ax.errorbar(LoadCont.load, LoadCont.friction, yerr=LoadCont.err, fmt='ks', capsize=3, label='Grafeno contaminado (carga)')
ax.errorbar(UnloadCont.load, UnloadCont.friction, yerr=UnloadCont.err, fmt='ks', fillstyle='none', capsize=3, label='Grafeno contaminado (descarga)')
ax.set_ylim(0,2.0)
ax.set_yticks(np.arange(0,2.5,0.5))
ax.set_yticklabels(np.arange(0,2.5,0.5), **a_font, fontsize=14)
ax.set_xlim(-10, 130)
ax.set_xticks(np.arange(0, 140, 20))
ax.set_xticklabels(np.arange(0,140,20), **a_font, fontsize=14)
ax.set_xlabel('Força normal (nN)', **a_font, fontsize=16)
ax.set_ylabel('Força lateral (nN)', **a_font, fontsize=16)
ax.legend(prop={'family' : 'Arial'})
plt.show()
```

```python
fig1.savefig('Load-Unload-Plot.png', dpi=300, bbox_inches='tight')
```

```python
fig2, (ax1, ax2) = plt.subplots(2, sharex=True)
ax1.errorbar(LoadFresh.load, LoadFresh.friction, yerr=LoadFresh.err, fmt='ko', capsize=3, label='Grafeno limpo (carga)')
ax1.errorbar(UnloadFresh.load, UnloadFresh.friction, yerr=UnloadFresh.err, fmt='ko', capsize=3, fillstyle='none', label='Grafeno limpo (descarga)')
ax1.set_ylim(0,2.0)
ax1.set_yticks(np.arange(0,2.5,0.5))
ax1.set_yticklabels(np.arange(0,2.5,0.5), **a_font, fontsize=14)
ax2.errorbar(LoadCont.load, LoadCont.friction, yerr=LoadCont.err, fmt='ks', capsize=3, label='Grafeno contaminado (carga)')
ax2.errorbar(UnloadCont.load, UnloadCont.friction, yerr=UnloadCont.err, fmt='ks', fillstyle='none', capsize=3, label='Grafeno contaminado (descarga)')
ax2.set_xlim(-10, 130)
ax2.set_xticks(np.arange(0, 140, 20))
ax2.set_xticklabels(np.arange(0,140,20), **a_font, fontsize=14)
ax2.set_xlabel('Força normal (nN)', **a_font, fontsize=16)
ax2.set_ylim(0,2.0)
ax2.set_yticks(np.arange(0,2.5,0.5))
ax2.set_yticklabels(np.arange(0,2.5,0.5), **a_font, fontsize=14)
plt.text(-30, 1.0, 'Força lateral (nN)', rotation=90, **a_font, fontsize=16)
ax1.legend(prop={'family' : 'Arial'}, loc='lower right')
ax2.legend(prop={'family' : 'Arial'})
ax1.text(-7.0, 1.7, "(a)", **a_font, fontsize=16)
ax2.text(-7.0, 1.7, "(b)", **a_font, fontsize=16)
plt.show()
```

```python
fig2.savefig('Load-Unload-2-Rows-1-Col.png', dpi=300, bbox_inches='tight')
```
